2. FIND OUT THE AGE DISTRIBUTION FROM THE DATASET

```python
plt.hist(df['Age'], bins=10, color='blue')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Age Distribution')
plt.xticks(np.arange(0,81,5))
plt.show()
```

3. RENAMING COLUMNS

```python
df.rename(columns= {'Food' : 'Food_Type'}, inplace=True)
df.rename(columns= {'Juice' : 'Drink_Type'}, inplace=True)
df.head(5)
```

4. REPLACING VALUES IN COLUMN

```python
df['Nationality'].value_counts()
```

```python
df['Nationality'] = df['Nationality'].replace({'MY':'Malaysian'})
df['Nationality'] = df['Nationality'].replace({'M':'Malaysian'})
df['Nationality'] = df['Nationality'].replace({'MY':'Malaysian'})
df.head(20)
```

5. SHOW CHAERT OF FAMALE AND MALE DESSERT PREFERENCE

```python
Gender_Nationality=pd.crosstab(df['Dessert'], df['Gender'])
explode_values=[0.05, 0, 0.05]
Gender_Nationality.plot(kind='pie', autopct="%0.1f%%", explode=explode_values, shadow=True, radius=0.9, subplots= True)
plt.title('Female and Male Dessert Preferrence')
plt.show()
```

6. FIND ALL THE DATA OF MALAYSIAN MEN
Gives you the Data of Malaysian men

```python
df[(df['Gender'] == 'Male') & (df['Nationality'].str.contains('Malaysia'))]
```

7. DOES TYPE OF FOOD EATEN AFFECT PREFERENCE TO DESSERT?
   Here we look at preference of dessert between traditional food and western food

```python
Food_by_Dessert=pd.crosstab(df['Food_Type'], df['Dessert'])
Food_by_Dessert.plot(kind='bar')
plt.xlabel('Food Type')
plt.ylabel('Counts')
plt.yticks(np.arange(0,101,5))
plt.title('Food by Dessert Preference')
plt.legend(title='Dessert preference', fontsize=8)
plt.show()
```
